```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
"true value 값을 이제는 placeholder를 사용하여 만든경우"
"세션 run 할 시 feed_dict=에 value를 꼭 넣어줘야 동작함."
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
"""placeholder(tf.float32, shape = [None])
    shape = [None] : 1차원 어레이 구조를 가짐 
    @Param None : 어느 값이 들어와도 문제 없음.
"""
```

```python
"tf.Variable(shape결정,이름 설정)"
"tf.random_normal([1]) : a와 b의 값을 모르기 때문에 랜덤한 1차원 배열을 생성"
W = tf.Variable(tf.random_normal([1]),name="weight")
b = tf.Variable(tf.random_normal([1]),name="bias")
```

```python
hypothesis = W * X + b
```

```python
"평균을 내주는 reduce_mean"
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
"minimize 과정 "
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
"Adam Optimazer"
optimizer_Adam = tf.train.AdamOptimizer(0.0001).minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
for step in range(2500):
    "리스트에 묶에서 실행 시킬수 있다. "
    """
    [cost,W, b, train] 각 실행 시키고 싶은 과정 
    """
    costValue, WValue,bValue, _ = sess.run([cost,W,b,optimizer_Adam],feed_dict={X:[1,2,3],Y:[2,4,5]})
    "sess.run(train)"
    if step % 20 == 0 :
        print(step, costValue, WValue, bValue)
```

# 구성한 모델 성능 테스트
