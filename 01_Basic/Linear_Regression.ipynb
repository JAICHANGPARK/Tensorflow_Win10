```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

# Tensor Graph 구현 과정 

```python
trainXdata = [1,2,3,4,5,6]
trainYdata = [10,2,3,5,1,7]
```

```python
"tf.Variable(shape결정,이름 설정)"
"tf.random_normal([1]) : a와 b의 값을 모르기 때문에 랜덤한 1차원 배열을 생성"
W = tf.Variable(tf.random_normal([1]),name="weight")
b = tf.Variable(tf.random_normal([1]),name="bias")
```

```python
hypothesis = trainXdata * W + b
```

```python
"평균을 내주는 reduce_mean"
cost = tf.reduce_mean(tf.square(hypothesis - trainYdata))
```

```python
"minimize 과정 "
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)
```

# 실행과 업데이트 과정

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
for step in range(2500):
    sess.run(train)
    if step % 20 == 0 :
        print(step, sess.run(cost), sess.run(W), sess.run(b))
```
