```python
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = [1,2,3,4,5,6]
Y = [2,5,3,7,1,9]
```

```python
W = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
hypothesis = W * X + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
w_Value = []
bias_Value = []
cost_Value = []
```

```python
for step in range(-30,50):
    feed_w = step * 0.1
    feed_b = step * 0.1
    currnetCost, currentW, currentB = sess.run([cost,W,b],feed_dict={W:feed_w, b: feed_b})
    w_Value.append(currentW)
    cost_Value.append(currnetCost)
    bias_Value.append(currentB)
    
```

```python
plt.figure()
plt.plot(w_Value, cost_Value)
plt.show()
```

```python
print(bias_Value)
print(w_Value)
print(cost_Value)

fig = plt.figure(1)
plot3ds = fig.gca(projection='3d')
plot3ds.plot_surface(bias_Value,w_Value,cost_Value)
plt.show()
```
