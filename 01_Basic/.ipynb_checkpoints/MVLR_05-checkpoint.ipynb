### MVLR (Multi Variable Linear Regression) in tensorflow}

```python
import tensorflow as tf
import time
```

```python
"""
x1_data = [79., 83., 90., 83., 89.]
x2_data = [92., 90., 88., 95., 70.]
x3_data = [82., 75., 92., 100., 70.]
y_data = [148., 130., 175., 198., 187.]
"""
xData = [[79.,92.,82.],[83.,90.,75.],[90.,88.,92],[83.,95.,100.],[89.,70.,70.]]
yData = [[148.],[130.],[175.],[198.],[187.]]
```

```python
"x의 n 값은 원하는 만큼 들어갈 수 있다고 본다."

X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
start = time.time()
for step in range(2001):
    cost_val, hy_val, _ = sess.run(
        [cost, hypothesis, train], feed_dict={X: xData, Y: yData})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPrediction:\n", hy_val)
end = time.time() - start
```

```python
print("수행 시간 : " + str(end))
```
