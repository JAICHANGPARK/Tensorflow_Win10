## 2. Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.array([-1.,1.,2.])
```

```python
X = np.arange(-5.0,5.0,0.1)
```

```python
sigmoid(x)
```

- Broadcast : numpy array와 scala value의 연산을 numpy array의 원소 각각과 스칼라값의 연산으로 바꿔 수행하는 것

```python
plt.figure(1)
plt.plot(X, sigmoid(X))
plt.ylim(-.1, 1.1)
plt.show()
```

## 3. ReLU

- [numpy.maximum](https://docs.scipy.org/doc/numpy/reference/generated/numpy.maximum.html) 
`Compare two arrays and returns a new array containing the element-wise maxima. If one of the elements being compared is a NaN, then that element is returned. If both elements are NaNs then the first is returned. The latter distinction is important for complex NaNs, which are defined as at least one of the real or imaginary parts being a NaN. The net effect is that NaNs are propagated.`

```python
# 0과 입력된 x의 값을 비교하여 0보다 큰 수를 출력
def relu(x):
    return np.maximum(0, x)
```

```python
plt.figure(2)
plt.plot(X, relu(X))
plt.ylim(-.1, 1.1)
plt.show()
```
