#module import하기 

```python
import tensorflow as tf
```

```python
node1 = tf.constant(1.0, dtype=tf.float32)
node2 = tf.constant(5.0)
#생성한 tensor 즉, node1, node2를 더한 새로운 tensor node3 생성
node3 = tf.add(node1,node2)
```

```python
print(node1)
print(node2)
print(node3)
```

```python
#Tensorflow는 Session에 넣어 동작을 시행해야 한다.
sess = tf.Session()
result = sess.run(node3)
print(sess.run([node1, node2]))
print(result)
```

#placeholder

###어떠한 값이 들어 있는지 모르는 tensor 노드를 추가한다. 

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
node_add = X + Y
```

```python
print(X)
print(Y)
```

```python
#feed_dict에서 값을 넣어 placeholder에 값을 넘겨준다. 
print(sess.run(node_add,feed_dict={X:3, Y: 5.5}))
print(sess.run(node_add,feed_dict={X:[1,3], Y: [2,2]}))
```
